```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk import word_tokenize
```

```python
f=open('corpusCNN.txt','rt',encoding='UTF8')
data=f.read()
data=data.replace("(", " ").replace(")"," ").replace("."," ").replace(","," ").replace("--"," ").replace("'"," ")
```

```python
stoken=sent_tokenize(data) #구문을 문장 단위로 분리
print(stoken)
```

```python
wtoken= word_tokenize(data) #문장을 단어 단위로 분리
print(wtoken)
```

```python
from nltk.tag import pos_tag
tagged_list = pos_tag(word_tokenize(data))
tagged_list
```

```python
freq_word = nltk.FreqDist(tagged_list)
print(freq_word.most_common(70))
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
word_token_mod = []
for w in tagged_list:
    if w not in stop_words:
        word_token_mod.append(w)

print(word_token_mod)
```

```python
freq_word = nltk.FreqDist(word_token_mod)
print(freq_word.most_common(70))
```
