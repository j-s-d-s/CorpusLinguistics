```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
def bbcUrl(search):
    url_temp = "https://www.bbc.co.uk/search?q={search}".format(search=search)
    driver = webdriver.Chrome("./chromedriver.exe")
    driver.get(url_temp)
    time.sleep(2)
    
    return bbcUrlList(driver)
```

```python
def bbcUrlList(driver):
    url_temp = driver.current_url
    for pageNo in range(5,15): 
        url = url_temp + "&page={pageNo}".format(pageNo=pageNo)
        #print(url)
        driver.get(url)
        time.sleep(2)
        
        bbcUrls_List = []        
        bbcUrls = driver.find_elements_by_css_selector(".css-1aofmbn-PromoHeadline a")
        
        for url in bbcUrls:
            bbcUrls_List.append(url.get_attribute('href'))
        
        return  bbcInfo(driver,bbcUrls_List)
```

```python
def bbcInfo(driver, bbcUrls_List):
    for urlList in bbcUrls_List:
        driver.get(urlList)
        time.sleep(2)
        #print(urlList)
        
        newsTitle =""
        newsTitleList = driver.find_elements_by_css_selector("#main-heading")
        if not newsTitleList:
            newsTitleList = driver.find_elements_by_css_selector(".headline")
        else:
            newsTitle_temp = newsTitleList[0]
            newsTitle = newsTitle_temp.text        
        print(newsTitle)
       

        newsContents = ""
        newsContentsList = driver.find_elements_by_css_selector(".css-14im5m3-ArticleWrapper")
        for newsContent in newsContentsList:
            newsContents = newsContents + newsContent.text
        print(newsContents)
        print('ㅡㅡㅡㅡㅡ 다음 ㅡㅡㅡㅡㅡ')
```
