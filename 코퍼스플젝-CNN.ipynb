```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
def cnnUrl(search):
    url_temp = "https://edition.cnn.com/search?q={search}&size=10".format(search=search)
    driver = webdriver.Chrome("./chromedriver.exe")
    driver.get(url_temp)
    time.sleep(2)
    
    return cnnUrlList(driver)
```

```python
def cnnUrlList(driver):
    url_temp = driver.current_url
    for pageNo in range(1,6): 
        url = url_temp + "&page={pageNo}".format(pageNo=pageNo) + "&type=article"
        #print(url)
        driver.get(url)
        time.sleep(2)
        
        cnnUrls_List = []        
        cnnUrls = driver.find_elements_by_css_selector(".cnn-search__result-headline a")
        
        for url in cnnUrls:
            cnnUrls_List.append(url.get_attribute('href'))
        
        return  cnnInfo(driver,cnnUrls_List)
```

```python
def cnnInfo(driver,cnnUrls_List):
    for urlList in cnnUrls_List:
        driver.get(urlList)
        time.sleep(2)
        #print(urlList)
        
        newsTitle =""
        newsTitleList = driver.find_elements_by_css_selector(".l-container .pg-headline")
        if not newsTitleList:
            newsTitleList = driver.find_elements_by_css_selector(".headline")
        else:
            newsTitle_temp = newsTitleList[0]
            newsTitle = newsTitle_temp.text        
        print(newsTitle)
       

        newsContents = ""
        newsContentsList = driver.find_elements_by_css_selector(".zn-body__paragraph")
        for newsContent in newsContentsList:
            newsContents = newsContents + newsContent.text
        print(newsContents)
        print('ㅡㅡㅡㅡㅡ 다음 ㅡㅡㅡㅡㅡ')
```
